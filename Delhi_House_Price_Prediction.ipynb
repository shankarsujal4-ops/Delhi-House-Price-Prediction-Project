```python
data.drop(columns = ['Unnamed: 0', 'Address','Lift', 'Landmarks', 'desc', 'Price_sqft'],inplace = True)
```

Filling missing Values

```python
data['Balcony'].value_counts()
```

```python
data['Balcony'] = data['Balcony'].apply(lambda x: 5 if pd.notnull(x) and x > 5 else x)
```

```python
data['Balcony'].value_counts()
```

```python
data['has_balcony'] = data['Balcony'].notnull().astype(int)
data['Balcony'] = data['Balcony'].fillna(data['Balcony'].median())
```

```python
data['Status'] = data['Status'].fillna('Unknown')
```

```python
data['Status'].value_counts()
```

```python
data['parking'].value_counts()
```

```python
data['parking'] = data['parking'].apply(lambda x: np.nan if x > 5 else x)
data['has_parking']=data['parking'].notnull().astype(int)
data['parking'] = data['parking'].fillna(0)
```

```python
data['has_parking'].value_counts()
```

```python
data['Furnished_status']=data['Furnished_status'].fillna('Unkown')
```

```python
data['Furnished_status'].value_counts()
```

```python
data.isna().sum()
```

Removing outliers on the basis of area per room

```python
(data['area']/data['Bedrooms']).describe()
```

```python
data = data[((data['area']/data['Bedrooms']) >= 300)]
data.describe()
```

Removing Price per square feet outliers by grouping latitude, longitude and bedrooms.


```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=0)
data['location_cluster'] = kmeans.fit_predict(data[['latitude', 'longitude']])  #Clustering latitudes and longitudes. This cluster resembles locality.
```

```python
data['price_per_sqft'] = data['price'] / data['area']
data['price_per_sqft'].describe()
```

```python
def remove_outliers(df):
  df_out = pd. DataFrame()
  for(cluster, bedroom), subdf in df.groupby(['location_cluster', 'Bedrooms']):
    m = subdf['price_per_sqft'].mean()
    s = subdf['price_per_sqft'].mean()
    reduced_df = subdf[(subdf['price_per_sqft'] > (m-s))  & (subdf['price_per_sqft'] > (m-s))]

    df_out = pd.concat([df_out, reduced_df], ignore_index=True)
  return (df_out)
data = remove_outliers(data)
```

```python
data.describe()
```

```python
data.drop(columns = ['price_per_sqft'], inplace = True)
```

```python
data.head()
```

Model Building

```python
X = data.drop(columns = ['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
Column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['Status', 'neworold', 'Furnished_status', 'type_of_building' ]),
    remainder='passthrough'
)
scaler = StandardScaler()
lr = LinearRegression()
pipe = make_pipeline(Column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
Column_trans.get_feature_names_out()
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

Applying Lasso


```python
lasso = Lasso()
pipe = make_pipeline(Column_trans, scaler, lasso)
pipe.fit(X_train, y_train)
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

Applying Ridge


```python
ridge = Ridge()
pipe = make_pipeline(Column_trans, scaler, ridge)
pipe.fit(X_train, y_train)
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

Random Forest Regressor


```python
from sklearn.ensemble import RandomForestRegressor
rand_forest = RandomForestRegressor()
pipe = make_pipeline(Column_trans, scaler, rand_forest)
pipe.fit(X_train, y_train)
y_pred_rand_forest = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_rand_forest)
```

```python
train_score = pipe.score(X_train, y_train)
test_score = pipe.score(X_test, y_test)

print(train_score, test_score)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipe, X, y, cv=5)
print(scores.mean())
```
